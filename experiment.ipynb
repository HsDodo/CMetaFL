```python
import pandas as pd


# Load the CSV file into a pandas DataFrame
df = pd.read_csv(f'./result/UCI-HAR-Result.csv')



# Define a window size for the rolling average
window_size = 10

# Apply the rolling window and calculate the mean for smoothing
smoothed_df = df.rolling(window=window_size).mean()

# Due to the rolling average, the first few values will be NaN, so we drop those rows
smoothed_df = smoothed_df.dropna()

# Display the first few rows of the smoothed dataframe to verify
smoothed_df.head()
# Display the first few rows of the dataframe to understand its structure
df = smoothed_df
df.head()
```

```python
import matplotlib.pyplot as plt

# Set the style to default for a white background
plt.style.use('default')


markers = ['o', 's', '^', 'v', 'D', '*', 'x']

# Plotting all methods except 'Mahal - Test/Acc (global)' and 'NoAttack - Test/Acc (global)'
i = 0
# colors = ['#8fcec9', '#ffbe7b', '#fa7f6f', '#81b0d2', '#beb8dc', '#999999', '#84a049']
for column in df.columns[1:]:
    plt.plot(df['round'], df[column], label=column, marker=markers[i], markersize=5, linewidth=1,markevery = 20)
    i = i + 1

# Adding the title and labels
plt.title(f'UCI-HAR dataset',fontsize=18)
plt.xlabel('Communication round',fontsize=18)
plt.ylabel('Test Accuracy',fontsize=18)

# Adding a legend
plt.legend()
plt.legend(loc='lower left',fontsize=14)
# plt.legend(loc='best',fontsize=12)

# Show a grid
plt.grid(True)
# 将plt保存
plt.gcf().savefig(f'./result/UCI-HAR_Result.png',format='png')
# plt.figure(figsize=(20, 6))

# Show the plot with a white background and adjusted line styles for visibility
plt.show()

```
